```python
def Infos_artist(Ddepart, Dfin):
    
    dt = timedelta(days=7)
    
    D = Ddepart
    
    artistes_ = []
    artistes = []
    titles_  = []
    titles   = []
    places   = ['1']
    billWeeks = []
    prevWeek = []         
    
    while D > Dfin:
       
        # pattern commun de l'url pour tous les titres
        URLbeg = 'https://www.billboard.com/charts/hot-100'
        
        # boucle le code pour obtenir informations sur chaque titre
        #for Titre in Stocks:
        #    URL = URLbeg + Titre   
        #    req = requests.get(URL)
        #    soup = BeautifulSoup(req.text, "lxml")
            
        req = requests.get(URLbeg)
        soup = BeautifulSoup(req.text, "lxml")
        
        # récupère le numéro 1
        recup = soup.find_all('div', class_='chart-number-one__title')
        titles_ += [str(title.text) for title in recup]   
        
        # manquera juste le numero 1
        recup = soup.find_all('span', class_='chart-list-item__title-text')
        titles_ += [str(title.text) for title in recup]
        
        recup = soup.find_all('div', class_='chart-list-item__rank')
        places += [str(place.text[1:-1]) for place in recup]
        
        # a voir plus tard les deux ensembles
        recup = soup.find_all('div', class_='chart-list-item__artist') # probleme
        artistes_ += [str(artiste.text) for artiste in recup]
        recup = soup.find_all('a') # probleme
        artistes_ += [str(artiste.text) for artiste in recup]
        
        recup = soup.find_all('div', class_="chart-list-item__weeks-on-chart") 
        billWeeks += [str(bill.text) for bill in recup]
        
        recup = soup.find_all('div', class_="chart-list-item__weeks-at-one") 
        prevWeek += [str(prW.text) for prW in recup]
        
        D -= dt
        
    for iloop in range(len(titles_)):
        titles.append(titles_[iloop].replace('\n','')) 
        
    for iloop in range(len(artistes_)):
        artistes.append(artistes_[iloop].replace('\n','')) 
          
    return [titles]
```

```python
Infos_artist(date(2018, 12, 1), date(2018, 11, 1))
```
